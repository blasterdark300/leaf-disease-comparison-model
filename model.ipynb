*Contoh Mengambil Beberapa Gambar Dataset*

```python
import os
from PIL import Image

# 1. Definisi Path
src_dir = r"C:\Users\crism\Downloads\archive\Rice Disease\Healthy"
dst_dir = r"C:\Users\crism\Downloads\daun3\archive\data\Rice_healthy"
target_size = (256, 256)
limit = 1100  # Batasi tepat 1100 gambar

# 2. Buat folder tujuan jika belum ada
if not os.path.exists(dst_dir):
    os.makedirs(dst_dir)
    print(f"Folder {dst_dir} berhasil dibuat.")

# 3. Proses Copy dan Resize dengan Limit
count = 0
valid_extensions = ('.jpg', '.jpeg', '.png', '.bmp')

print(f"Sedang memproses maksimal {limit} gambar...")

# Ambil daftar file dan urutkan agar prosesnya konsisten
filenames = [f for f in os.listdir(src_dir) if f.lower().endswith(valid_extensions)]

for filename in filenames:
    if count >= limit:
        break  # Berhenti jika sudah mencapai 1000
    
    try:
        img_path = os.path.join(src_dir, filename)
        save_path = os.path.join(dst_dir, f"rice_healthy_{count+1}.jpg") # Rename agar rapi

        with Image.open(img_path) as img:
            # Resize dengan kualitas tinggi
            img_resized = img.resize(target_size, Image.Resampling.LANCZOS)
            
            # Konversi ke RGB (menghindari error channel)
            if img_resized.mode != "RGB":
                img_resized = img_resized.convert("RGB")
            
            img_resized.save(save_path, "JPEG", quality=90)
            count += 1
            
            if count % 100 == 0:
                print(f"Berhasil memproses {count} gambar...")
    
    except Exception as e:
        print(f"Gagal memproses {filename}: {e}")

print("-" * 30)
print(f"Selesai! Total {count} gambar 'Rice_healthy' siap di {dst_dir}")
```

*Undersampling*

```python
import os
import shutil
import numpy as np

dataset_path = r"C:\Users\crism\Downloads\daun3\daun\dataset"
backup_path = r"C:\Users\crism\Downloads\daun3\daun\dataset_unused"

# 1. Hitung jumlah gambar di tiap kelas dulu
class_counts = []
for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)
    if os.path.isdir(class_folder):
        class_counts.append(len(os.listdir(class_folder)))

# 2. Tentukan ambang batas (Threshold) pakai Median agar lebih adil
threshold = int(np.median(class_counts)) 
print(f"Ambang batas otomatis berdasarkan Median: {threshold}")

# 3. Eksekusi pemindahan
if not os.path.exists(backup_path):
    os.makedirs(backup_path)

for class_name in os.listdir(dataset_path):
    class_folder = os.path.join(dataset_path, class_name)
    if os.path.isdir(class_folder):
        images = os.listdir(class_folder)
        if len(images) > threshold:
            print(f"Memotong {class_name}: {len(images)} -> {threshold}")
            to_remove = images[threshold:]
            
            dest_folder = os.path.join(backup_path, class_name)
            os.makedirs(dest_folder, exist_ok=True)
            
            for img in to_remove:
                shutil.move(os.path.join(class_folder, img), os.path.join(dest_folder, img))

print("Proses penyeimbangan selesai!")
```

*Split Dataset*

```python
import os
import shutil
import random

# Path asli dan path tujuan split
root_dir = r"C:\Users\crism\Downloads\daun3\daun\dataset"
base_dir = r"C:\Users\crism\Downloads\daun3\daun\dataset_final"

classes = os.listdir(root_dir)

# Membuat folder utama
for split in ['train', 'val', 'test']:
    for cls in classes:
        os.makedirs(os.path.join(base_dir, split, cls), exist_ok=True)

# Proses pembagian
for cls in classes:
    src_path = os.path.join(root_dir, cls)
    all_files = os.listdir(src_path)
    random.shuffle(all_files) # Acak gambar agar distribusi merata
    
    # Hitung jumlah file per split
    train_idx = int(len(all_files) * 0.8)
    val_idx = int(len(all_files) * 0.9)
    
    train_files = all_files[:train_idx]
    val_files = all_files[train_idx:val_idx]
    test_files = all_files[val_idx:]
    
    # Fungsi copy
    for file in train_files:
        shutil.copy(os.path.join(src_path, file), os.path.join(base_dir, 'train', cls, file))
    for file in val_files:
        shutil.copy(os.path.join(src_path, file), os.path.join(base_dir, 'val', cls, file))
    for file in test_files:
        shutil.copy(os.path.join(src_path, file), os.path.join(base_dir, 'test', cls, file))

print(f"Selesai! Dataset telah di-split 80:10:10 di {base_dir}")
```
